```python
from bs4 import BeautifulSoup
import requests
```

```python
website = 'http://automationpractice.com/index.php?id_category=3&controller=category'
```

```python
response = requests.get(website)
```

```python
#Check status code
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
#Find By Id
contact_us = soup.find(id='contact-link').find('a').get_text()
contact_us
```

```python
#Find By Class, Find method only return one element and this is the first
product = soup.find(class_='ajax_block_product').find(class_='content_price')
```

```python
#Find all method by id or class, this method return a list
results_ = soup.findAll(class_='ajax_block_product')
```

```python
len(results_)
```

```python
#Select_one method return the first element with css class selector
soup.select_one('.shop-phone').get_text().strip().replace('Call us now: ', '')
```

```python
#Get the contact link
soup.select_one('#contact-link').get_text().strip()
```

```python
#Find By Attribute
soup.find(itemprop='name').find('a').get('title')
```

```python
#select product one and get next sibling
soup.select_one('.ajax_block_product').find_next_sibling().select_one('.product-name').get('title')
```

```python
#Go to product #7 and get previous sibling, from it get the price
soup.select('.ajax_block_product')[6].find_previous_sibling().select_one('.product-name').get('title')
```

```python
#Find Elements parent
soup.select('.ajax_block_product')[6].find_parent()
```

```python
#Get a single link
soup.select_one('#contact-link').find('a').get('href')
```

```python
#Find a multiple links
nameElements = soup.findAll(itemprop='name')

links = []

for element in nameElements:
    links.append(element.find('a').get('href'))
    
links
```
